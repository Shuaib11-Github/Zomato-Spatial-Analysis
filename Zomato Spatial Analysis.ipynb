```python
import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Zomato.csv')
df
```

```python
df.isna().sum()
```

```python
df.dropna(axis=0, subset=['location'], inplace=True)
```

```python
df.isna().sum()
```

```python
len(df['location'].unique())
```

```python
locations = pd.DataFrame()
```

```python
locations['Name'] = df['location'].unique()
locations
```

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent= 'app')
```

```python
lat = []
lon = []
for location in locations['Name']:
    location = geolocator.geocode(location)
    if location is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(location.latitude)
        lon.append(location.longitude)
```

```python
locations['lat'] = lat
locations['lon'] = lon
```

```python
locations.head()
```

```python
locations.to_csv('Zomato_locations.csv', index = False)
```

#### We have found out latitude and longitude of each location listed in the dataset using geopy
#### This is used to plot maps.

```python
Rest_locations = df['location'].value_counts().reset_index()
Rest_locations
```

```python
Rest_locations.columns = ['Name', 'Count']
Rest_locations
```

#### now combine both the dataframes

```python
Restaurant_locations = Rest_locations.merge(locations, on = 'Name', how = 'left').dropna()
Restaurant_locations
```

```python
df['rate'].unique()
```

```python
df.dropna(axis = 0, subset = ['rate'], inplace = True)
```

```python
def split(x):
    return x.split('/')[0]
```

```python
df['rating'] = df['rate'].apply(split)
```

```python
df.head()
```

```python
df['rating'].unique()
```

```python
df.replace('NEW', 0, inplace = True)
```

```python
df.replace('-', 0, inplace = True)
```

```python
df['rating'].unique()
```

```python
df.dtypes
```

```python
df['rating'] = pd.to_numeric(df['rating'])
```

```python
df.groupby('location')['rating'].mean().sort_values(ascending = False)
```

```python
avg_rating = df.groupby('location')['rating'].mean().sort_values(ascending = False).values
```

```python
location = df.groupby('location')['rating'].mean().sort_values(ascending = False).index
```

```python
rating = pd.DataFrame()
```

```python
lat = []
lon = []
for loc in location:
    loc = geolocator.geocode(loc)
    if loc is None:
        lat.append(np.nan)
        lon.append(np.nan)
    else:
        lat.append(loc.latitude)
        lon.append(loc.longitude)
```

```python
rating['location'] = location
rating['lat'] = lat
rating['lon'] = lon
rating['avg_rating'] = avg_rating
```

```python
rating.head()
```

```python
rating.isna().sum()
```

```python
rating.dropna(inplace = True)
```

```python
HeatMap(rating[['lat', 'lon', 'avg_rating']]).add_to(basemap)
```

```python
basemap
```

```python
filter = df['cuisines'] == 'North Indian'
df2 = df[filter]
df2.head()
```

```python
north_india = df2.groupby('location')['url'].count().reset_index()
north_india.columns = ['Name', 'count']
north_india
```

```python
north_india = north_india.merge(locations, on = 'Name', how = 'left').dropna()
north_india
```

#### Heatmap of North Indian restaurants

```python
HeatMap(north_india[['lat', 'lon', 'count']], zoom = 20, radius = 15).add_to(basemap)
basemap
```

```python
df['cuisines'].unique()
```

```python
def HeatMap_zone(zone):
    df2 = df[df['cuisines'] == zone]
    df_zone = df2.groupby('location')['url'].count().reset_index()
    df_zone.columns = ['Name', 'count']
    df_zone = df_zone.merge(locations, on = 'Name', how = 'left').dropna()
    HeatMap(df_zone[['lat', 'lon', 'count']], zoom = 20, radius = 15).add_to(basemap)
    return basemap
```
